```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the data
df = pd.read_excel("MLS-Analytics.xlsx", engine='openpyxl')
df.head()
```

```python
# add 0 instead of the nan values
df = df.replace(np.nan,0)
df.head()
```

```python
# Combine the columns Director Sports performacne Science, Head of Sports Science and Director of Sports Performance.
sum_of_columns = df["Director Sports Performance Science"]+ df["Head of Sports Science"] + df["Director of Sports Performance"]
df['Sports Performance Science Total'] = sum_of_columns
df.head()
```

```python
# drop the columns Director Sports Performance Science, Head of Sports Science and Director of Sports Performance
df = df.drop(["Director Sports Performance Science","Head of Sports Science","Director of Sports Performance"],axis=1)
df.head()
```

```python
# swap columns
columns_titles = ['Teams',
 'Performance Analyst',
 'Director of Soccer Analytics',
 'Sports Scientist',
 'Video/Data Analyst',
 'Performance Specialist',
 'Data Analyst',
 'Assistant GM',
 'Assistant Coach',
 'Data Science',
 'General Manager',
 'Director of Scouting',
 'Technical Director','Sports Performance Science Total','Division']
df=df.reindex(columns=columns_titles)
df.head()
```

```python
# Change column Sports Performance Science Total to Sports Performance Science
df=df.rename(columns = {'Sports Performance Science Total':'Head of Sports Performance Science'})

# Change RSL team name
df = df.replace({'Real Salt Lake City': 'Real Salt Lake'}, regex=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# change all floats to integers
columns_convert_lst = ['Head of Sports Performance Science', 'Performance Analyst',
       'Director of Soccer Analytics', 'Sports Scientist',
       'Video/Data Analyst', 'Performance Specialist', 'Data Analyst',
       'Assistant GM', 'Assistant Coach', 'Data Science', 'General Manager',
       'Director of Scouting', 'Technical Director']
df[columns_convert_lst] = df[columns_convert_lst].applymap(np.int64)
df.head()
```

```python
# Sort by team name
df = df.sort_values("Teams")
df.head()
```

```python
# create row sum column
df['Total Staff'] = df.sum(axis=1)
df = df[['Teams', 'Head of Sports Performance Science', 'Performance Analyst',
       'Director of Soccer Analytics', 'Sports Scientist',
       'Video/Data Analyst', 'Performance Specialist', 'Data Analyst',
       'Assistant GM', 'Assistant Coach', 'Data Science', 'General Manager',
       'Director of Scouting', 'Technical Director',
       'Total Staff','Division']]
df.head()
```

```python
# Total Staff MLS
df_total_staff = df.sort_values("Total Staff",ascending=False)
plt.figure(figsize=(20,15))
sns.set(font_scale=1.5)
ax = sns.barplot(y="Teams", x= "Total Staff", data=df_total_staff)
plt.xlabel("Teams",fontsize=20)
plt.ylabel("Jobs",fontsize=20)
plt.title("Analytics Staff by Teams in MLS", fontsize=20)
```

```python
df_total_staff["Total Staff"].mean()
```

```python
# Group by East and West
df_east_west = df.groupby(["Division"]).sum()
df_east_west
```

```python
# plot east and west
plt.figure(figsize=(20,15))
sns.set(font_scale=1.5)
sns.barplot(data= df_east_west, y = "Total Staff", x = df_east_west.index )
plt.xlabel("Division",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Total Staff by Division", fontsize=20)
```

```python
# plot east and west performance analysts
plt.figure(figsize=(20,15))
sns.set(font_scale=1.5)
sns.barplot(data= df_east_west, y = "Performance Analyst", x = df_east_west.index )
plt.xlabel("Division",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Total Performance Analysts by Division", fontsize=20)
```

```python
# plot performance analysts throughout the league
# Sort by team name
df_teams_pa = df.sort_values("Performance Analyst", ascending=False)
plt.figure(figsize=(20,15))
sns.set(font_scale=1.5)
sns.barplot(data= df_teams_pa, x = "Performance Analyst", y = "Teams" )
plt.xlabel("Division",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Total Performance Analyst Staff by Division", fontsize=20)
```

```python
# select numeric columns and calculate the sums
sums = df.select_dtypes(pd.np.number).sum().rename('total')

# append sums to the data frame
df_col_sums = df.append(sums)
df_col_sums.head()
```

```python
last_row = df_col_sums.iloc[[-1]]
# We take the transpose
last_row_df = last_row[['Head of Sports Performance Science', 'Performance Analyst',
       'Director of Soccer Analytics', 'Sports Scientist',
       'Video/Data Analyst', 'Performance Specialist', 'Data Analyst',
       'Assistant GM', 'Assistant Coach', 'Data Science', 'General Manager',
       'Director of Scouting', 'Technical Director']].T

last_row_df = last_row_df.sort_values("total", ascending=False)
last_row_df
```

```python
plt.figure(figsize=(20,15))
sns.set(font_scale=1.5)
sns.barplot(data= last_row_df, y = last_row_df.index, x = last_row_df.total )
plt.xlabel("Division",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Job Position related to Data in MLS", fontsize=20)
```
